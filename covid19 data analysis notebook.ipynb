### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries = list(corona_dataset_aggregated.index)
max_inf_rate = []

for country in countries:
    max_inf_rate.append(corona_dataset_aggregated.loc[country].diff().max())
#max_inf_rate 
corona_dataset_aggregated["max_inf_rate"] = max_inf_rate
```

```python
corona_dataset_aggregated.head()
```

### Task 3.4: create a new dataframe with only needed column 

```python
inf_rate_data = pd.DataFrame(corona_dataset_aggregated["max_inf_rate"])
```

```python
inf_rate_data.head()
```
